The frontend consists of three major operations.
Lexical analysis, syntax analysis, and semantic analysis.

 In contrast to prior compilers, which had separate programs for each of the procedures, 
These three processes are generally merged into modern parsers.

**Lexical analysis.**

The lexical analyzer divides a string indicating a program into tokens. 
This process is known as tokenization. These tokens are obtained using the programming language's rules and keywords. 
The Lexical analyzer returns the tokenized program after tokenization is complete.

**Syntax analysis/ Parser**

The Parser, also known as Syntax Analysis, takes the Lexical Analysis result and generates a syntax tree.
This is accomplished by taking each token obtained from the Lexical Analyzer one at a time 
and constructing the syntax tree using Context-Free Grammar.
This syntax tree, also known as a parse tree or derivation tree,
is built to detect ambiguity in the specific grammar. 

The syntax tree adheres to the following guidelines:

	1. Any identifier is considered an expression.
	2. An expression can be any number.
	3. Any operation performed in the supplied expression will always result in an expression.

 It checks whether the given program is semantically consistent with the language definition 
using a syntax tree and a symbol table.
 It collects type information and saves it in a syntax tree or a symbol table.
