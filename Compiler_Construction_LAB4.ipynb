# 1. Using an example or illustration, explain what is predictive parsing?
A predictive parser can be described as a top down parser that does not require back tracking when deciding which terminal should replace a non terminal. This works by making the choice oh the rule to be expanded to be made upon the next terminal symbol.

**Example**

E -> E + T | T

T -> T * F | F

F -> (E) | id

In order to create a predictive parser from the following language the following steps must be followed.

**1.Eliminate left recursion**

This grammar has left recursion since there are two productions whose right hand side begin with the same Non - terminal as the left hand side

E -> E + T | T

T -> T * F | F

Now eliminating left recursion

E -> TE’  E’ -> +TE’ | ε

T -> FT’  T’ -> *FT’ | ε

F -> (E) | id

**2.Check if left factoring is possible or not**

This is simply grouping of productions that begin with a similar terminal/non terminal. Similar to the grouping in algebraic equations(ab + ac becomes a(b + c)).
Since there are no such productions in this example this step is skipped

**3.Find First() and Follow() functions**

***First()***

First(E) -> First(T) -> First(F) -> { (,id } [ This means that the First() of E,T and F are the same ]

First(E’) -> { +,ε }

First(T’) -> { *,ε }


***Follow()***

Follow(E) -> { $,) } [ The start symbol automatically has a $ sign for a Follow() ]

Follow(E’) -> { $,) }

Follow(T) -> { +,$,) }

Follow(T’) -> { +,$,) }

Follow(F) -> { *,+,$,) }
